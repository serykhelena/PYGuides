# Свёрточные сети

На сегодняшний день глубокое обучение и нейронные сети используются почти везде. Например, их используют при разработке беспилотников, в различных системах компьютерного зрения, рекоммендательных системах и т.д. 

## Что такое нейронная сеть? 

Нейронная сеть названа так по аналогии с нервной системой у человека. Нейронная сеть обладает очень похожей структурой.

Но умные люди пишут, что искусственные нейроны работают точно также как нейроны в человеческой нервной системе, то есть собирают все входные сигналы и обрабатывают их. А дальше, передают свой выход на следующие нейроны (нейроны следующего слоя), с которыми они соединены. 

Нейронная сеть разделена на слои 3-х типов: 

* Входной слой. Исходные данные подаются на этот слой. 

* Скрытые слои. Это промежуточные слои между входным и выходным слоями, которые помогают сети находить сложные зависимости в данных, тем самым обучаться. 

* Выходной слой. Например, в случае задачи классификации с 5 классами, выходной слой будет состоять из 5-ти нейронов (каждый нейрон соответствует своему классу). 

Если посмотреть отдельно на работу одного нейрона, то: 

* $x_1, x_2, ...$ - это входы нейрона. Эти значения могут быть как исходной информацией, так и промежуточной между слоями

* $b_i$ - это блок биаса, то есть это такая константа, которая передаётся в функцию активации

* $w_0, w_1, ...$ - это значения весов

* $f$ - функция активации, она позволяет сделать нейронную сеть очень гибкой и помогает сети в целом обучаться. Существует большое количество разных функций активации, в целом можете и свою написать, но посмотрите сначала на те, которые уже существуют, их реально много. 

## Типы нейронных сетей 

Типы нейронных сетей разделяются на базе из архитектур:

1. Искуственные нейронные сети (ИНН = Artificial Neural Network = ANN): они состоят из множества нейроном (перцептронов) на каждом слое. ИНН также называют Feed-Forward Neural Network (нейронная сеть с прямой связью), потому что входные данные обрабатываются только в прямом направлении. 

2. Многослойный перцептрон (Multi-layered Perceptron): такие нейронные сети имеют 3 и более слоёв (то есть 1 входной слой, хотя бы 1 скрытый слой и 1 выходной слой). 

3. Рекурентные нейронные сети (Recurrent Neural Networks = RNN/LSTM)

4. Свёрточные нейронные сети (Convolutional Neural Networks = CNN): такие сети используют ядро/фильтры, чтобы достать фичи из данных, эта операция называет свёртка. Такие сети очень популярны при решении задач, связанных с обработкой изображений.
