## Как работает сегментация? 

Мы можем разделить (разбить) изображение на разные части, которые называются **сегментами**. 

Обрабатывать полностью картинку может быть плохой идеей, так как на ней могут быть области, в которых совсем нет полезной информации или просто вся картинка может быть очень большой и для её обработки понадобится прям очень много вычислительных мощностей. 

Разделив картинку на сегменты, мы можем использовать только важные части для обработки. Вот в этом то и есть весь смысл сегментации. 

Изображение - это набор различных пикселей. Мы можем сгруппировать их по каким-то правилам/особенностям при помощи сегментации. 

<p align="center">
    <img src="https://github.com/serykhelena/PYGuides/blob/main/notebooks/assets/cv_22.jpg?raw=true" alt="drawing" width="600"/>
</p>

Детекция объектов создаёт ограничивающую рамку (Bounding box) вокруг каждого объекта, но при этом ничего не говорит о форме объекта. Мы получаем только координаты прямоугольника, внутри которого находится объект. Но, а что, если мы хотим больше информации? 

Сегментация изображения создаёт пиксельную маску для каждого класса объектов. Этот метод даёт нам гораздо более детальное понимание объекта (ов) на изображении. 

Но зачем оно нам вообще? Зачем нам нужна эта пиксельная маска? Ведь работать с координатами прямоугольника намного проще, всего 4 точки? Давайте рассмотрим такой пример, чтобы ответить на эти логичные вопросы. 

## Зачем вообще нужна сегментация? 

Рак уже давно считается смертельной болезнью, и, к сожалению, люди всё ещё не научились полностью излечивать людей с таким диагнозом. Но чем раньше мы обнаружим его, тем больше шансов его вылечить. Максимально быстрое обнаружение раковых клеток может спасти множество жизней. 

Форма раковых клеток играет очень важную роль в определении степени тяжести заболевания. И в этом случае только детекции клеток явно недостаточно. 

Сегментация изображения в этом случае помогает подойти к этой проблеме более детально и получить более значимые результаты.

## Типы сегментации

Сегментацию изображений можно грубо разделить на 2 категории: 

* Семантическая сегментация (Semantic Segmentation)
* Сущностная сегментация (Instance Segmentation)

<p align="center">
    <img src="https://github.com/serykhelena/PYGuides/blob/main/notebooks/assets/cv_1.webp?raw=true" alt="drawing" width="600"/>
</p>

* На картинке слева каждый пиксель относится к определённому классу (задний фон или человек). Также все пиксели, которые относятся к одному классу, представлены одиним цветов (задний фон - чёрным цветом, люди - розовым). 

* На картинке справа каждый пиксель также как и в семантической сегментации относится к определённому классу, но разные обьъекты одного и того же класса представлены разными цветами (человек 1 - красный, человек 2 - зелёный, задний фон - чёрный и т.д.)

Если на картинке есть 5 людей, семантическая сегментация сфокусируется на классификации всех людей как единого экзмепляра (сущности). При это сущностная сегментация будет определять каждого из этих людей по отдельности.
